```python
# Importar bibliotecas necessárias
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Ler o dataset do arquivo CSV
# Suponha que o arquivo se chame "churn_data.csv"
churn_df = pd.read_csv("../../data/processed/telecom_churn_clean.csv")
```

# Selecionar a variável dependente e as variáveis independentes

```python
y = churn_df["churn"].values
X = churn_df[["account_length", "customer_service_calls"]].values
```

# Criar um classificador KNN 

```python
knn = KNeighborsClassifier(n_neighbors=6)
```

# Ajustar o classificador aos dados

```python
knn.fit(X, y)
```

# Suponha que X_new é um novo conjunto de dados para predizer


```python
X_new = [[130, 3], [200, 1]]
```

# Fazer previsões para X_new

```python
y_pred = knn.predict(X_new)
```

# Imprimir as previsões

```python
print("Predictions: {}".format(y_pred))
```
